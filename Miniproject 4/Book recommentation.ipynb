## NLP and clustering

```python
# Text Preprocessing

import pandas as pd
import numpy as np
import nltk
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Download necessary NLTK packages
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load dataset (ensure "Description" column exists)
df = pd.read_csv(r"C:\Users\Mohan\PycharmProjects\pythonProject\Miniproject4\Audible_Catalog_Cleaned.csv")

# Function to clean text
def clean_text(text):
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    text = text.strip()  # Remove whitespace
    text = " ".join([word for word in text.split() if word not in stopwords.words("english")])  # Remove stopwords
    return text

# Apply text cleaning to the "Description" column
df["cleaned_description"] = df["Description"].apply(clean_text)
```

```python
# Convert Text to Numerical Features

# Convert descriptions to numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)  # Limit to top 5000 words
X = vectorizer.fit_transform(df["cleaned_description"])
```

```python
# Apply Clustering (K-Means)

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for k in range(1, 11):  # Checking for clusters from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker="o", linestyle="--")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS (Within Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal Clusters")
plt.show()
```

```python
#Train K-Means with Optimal Clusters

optimal_k = 5  # Choose based on elbow curve
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)

# Display sample results
df[["Book Name", "Author", "Cluster"]].head(10)
```

```python
# Visualize Clustering Results

plt.figure(figsize=(12, 6))
sns.countplot(x=df["Cluster"], palette="coolwarm")
plt.xlabel("Cluster")
plt.ylabel("Number of Books")
plt.title("Book Clusters Based on Description Similarity")
plt.show()
```

```python
#Sample Book Recommendations Based on Clustering

def recommend_books(book_name, df, n=5):
    # Check if the book exists in the dataset
    if book_name not in df["Book Name"].values:
        return f"'{book_name}' not found in the dataset. Please check the spelling or choose another book."

    # Get the cluster of the input book
    book_cluster = df[df["Book Name"] == book_name]["Cluster"].values[0]
    
    # Get books in the same cluster (excluding the input book)
    recommendations = df[df["Cluster"] == book_cluster].copy()
    recommendations = recommendations[recommendations["Book Name"] != book_name]
    
    # If fewer books are available, return all of them
    if recommendations.shape[0] < n:
        return recommendations[["Book Name", "Author"]]
    
    return recommendations.sample(n)[["Book Name", "Author"]]

# Example Usage
book_to_recommend = "101 Amazing Facts About the Royal Family"  # Replace with a book from your dataset
print(recommend_books(book_to_recommend, df))

```

**Recommendation System Development**

```python
# Content-Based Filtering

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create TF-IDF matrix
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["Description"])  

# Compute Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to recommend books based on content similarity
def recommend_books_content(book_name, df, n=5):
    if book_name not in df["Book Name"].values:
        return f"'{book_name}' not found in the dataset."

    # Get book index
    book_idx = df[df["Book Name"] == book_name].index[0]

    # Get similarity scores & sort by highest
    sim_scores = list(enumerate(cosine_sim[book_idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n+1]

    # Get book recommendations
    book_indices = [i[0] for i in sim_scores]
    return df.iloc[book_indices][["Book Name", "Author"]]

# Example Recommendation
print(recommend_books_content("101 Amazing Facts About the Royal Family", df))
```

```python
# Clustering-Based Recommendations

from sklearn.cluster import KMeans

# Choose number of clusters
num_clusters = 10  

# Apply KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df["Cluster"] = kmeans.fit_predict(tfidf_matrix)

# Function to recommend books from the same cluster
def recommend_books_cluster(book_name, df, n=5):
    if book_name not in df["Book Name"].values:
        return f"'{book_name}' not found in the dataset."

    # Get cluster of the given book
    book_cluster = df[df["Book Name"] == book_name]["Cluster"].values[0]

    # Get books from the same cluster (excluding input book)
    recommendations = df[df["Cluster"] == book_cluster].copy()
    recommendations = recommendations[recommendations["Book Name"] != book_name]

    return recommendations.sample(n)[["Book Name", "Author"]] if len(recommendations) > 0 else "No recommendations found."

# Example Recommendation
print(recommend_books_cluster("101 Amazing Facts About the Royal Family", df))
```

```python
# Hybrid Approach

def hybrid_recommendation(book_name, df, n=5):
    content_recs = recommend_books_content(book_name, df, n)
    cluster_recs = recommend_books_cluster(book_name, df, n)

    # Combine recommendations, remove duplicates
    hybrid_recs = pd.concat([content_recs, cluster_recs]).drop_duplicates().head(n)
    
    return hybrid_recs if not hybrid_recs.empty else "No recommendations found."

# Example Recommendation
print(hybrid_recommendation("101 Amazing Facts About the Royal Family", df))
```

```python
# Model Evaluation: Precision, Recall, and RMSE

from sklearn.metrics import precision_score, recall_score, mean_squared_error
import numpy as np

# Simulated user ratings (1-5 scale)
df["User_Rating"] = np.random.randint(1, 6, df.shape[0])  

# Sample Predictions (Replace these with actual model predictions)
df["Content_Pred"] = df["User_Rating"] + np.random.normal(0, 0.5, df.shape[0])
df["Cluster_Pred"] = df["User_Rating"] + np.random.normal(0, 0.6, df.shape[0])
df["Hybrid_Pred"] = (df["Content_Pred"] + df["Cluster_Pred"]) / 2  # Hybrid combines both

# Calculate RMSE for each model
rmse_content = np.sqrt(mean_squared_error(df["User_Rating"], df["Content_Pred"]))
rmse_cluster = np.sqrt(mean_squared_error(df["User_Rating"], df["Cluster_Pred"]))
rmse_hybrid = np.sqrt(mean_squared_error(df["User_Rating"], df["Hybrid_Pred"]))

# Convert ratings to binary relevance (1 if rating ≥ 4, else 0)
true_labels = (df["User_Rating"] >= 4).astype(int)
content_preds = (df["Content_Pred"] >= 4).astype(int)
cluster_preds = (df["Cluster_Pred"] >= 4).astype(int)
hybrid_preds = (df["Hybrid_Pred"] >= 4).astype(int)

# Calculate Precision & Recall for each model
precision_content = precision_score(true_labels, content_preds)
recall_content = recall_score(true_labels, content_preds)

precision_cluster = precision_score(true_labels, cluster_preds)
recall_cluster = recall_score(true_labels, cluster_preds)

precision_hybrid = precision_score(true_labels, hybrid_preds)
recall_hybrid = recall_score(true_labels, hybrid_preds)

# Print Evaluation Results
print(f"Model Comparison:")
print(f"{'='*40}")
print(f"{'Model':<20}{'RMSE':<10}{'Precision':<10}{'Recall'}")
print(f"{'='*40}")
print(f"{'Content-Based':<20}{rmse_content:.2f}{precision_content:.2f}{recall_content:.2f}")
print(f"{'Clustering':<20}{rmse_cluster:.2f}{precision_cluster:.2f}{recall_cluster:.2f}")
print(f"{'Hybrid':<20}{rmse_hybrid:.2f}{precision_hybrid:.2f}{recall_hybrid:.2f}")
```
