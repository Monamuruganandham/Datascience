```python
import pandas as pd
customers=pd.read_csv(r"customers_dfcleaned.csv")
customers.info()
```

```python
import pandas as pd
import numpy as np
customers = customers.replace({np.nan:None})
customers

```

```python
query="drop table customers_table"
cursor.execute (query)
```

```python
# Create a string for the table columns
column_definitions = ",".join(
    f"{i} {j}"
    for i, j in zip(customers.columns, customers.dtypes)
).replace("float64", "FLOAT").replace("object", "VARCHAR(255)").replace("int64", "INT")

# Define the table name and create the table with an auto-incrementing id
table_name = "customers_table"
create_table_query = f"""
CREATE TABLE if not exists {table_name} (
    id INT AUTO_INCREMENT PRIMARY KEY,
    {column_definitions}
)
"""


cursor.execute(create_table_query)

print("table created successfully")
```

```python
query=f"describe {table_name}"
cursor.execute(query)
for i in cursor:
    print(i)
```

```python
# Prepare the INSERT query
insert_query = f"""
INSERT INTO {table_name} ({', '.join(customers.columns)})
VALUES ({', '.join(['%s'] * len(customers.columns))})
"""

# Execute the query for each row in the DataFrame
with connection.cursor() as cursor:  # Create a cursor object
    for i in customers.itertuples(index=False, name=None):  # Iterate over each row
        cursor.execute(insert_query, i)  # Use the row as the values for the query
    connection.commit()  # Commit the transaction

# Close the connection
connection.close()
print("Data inserted successfully")
```
