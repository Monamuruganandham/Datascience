MERGED DATA

```python
customers=pd.read_csv(r"customers_dfcleaned.csv")
products=pd.read_csv(r"products_dfcleaned.csv")
stores=pd.read_csv(r"stores_dfcleaned.csv")
sales=pd.read_csv(r"sales_dfcleaned.csv")
exchange_rates=pd.read_csv(r"exchangerates_dfcleaned.csv")
```

```python
#merge sales with product key
merged_df = sales.merge(products, on='ProductKey', how='left')
#merge mergeddf with customerkey
merged_df =merged_df.merge(customers,on='CustomerKey',how='left')
#merge mergeddf with 
merged_df=merged_df.merge(stores,on='StoreKey',how='left')



```

```python
merged_df
```

```python
print("Merged DataFrame Columns:", merged_df.columns)
print("Exchange Rates Columns:", exchange_rates.columns)
```

```python

final_df = pd.merge(merged_df, exchange_rates,left_on=['Order_Date','Currency_Code'],right_on=['Date','Currency'], how="inner")



```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
duplicates=final_df.duplicated().sum()
print(f"number of duplicates found:{duplicates}")
```

```python
final_df.to_csv("finalmergeddf.csv",index=False)
```
